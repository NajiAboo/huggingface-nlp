```python
from transformers import pipeline
```

```python
classifer = pipeline("sentiment-analysis")
```

```python
classifer("I love to teach")
```

```python
classifer(["I love to code","I dont want to sit idle"])
```

```python
txtgen = pipeline("text-generation")
```

```python
txtgen(" I love to teach about NLP ")
```

```python
masker = pipeline("fill-mask")
```

```python
masker("I am teaching <mask> and it will help computers to undersand the human language")
```

```python
ner = pipeline("ner")
```

```python
ner("My name is Naji Aboo and I work at Triassic Solutions", grouped_entities=True)
```

```python
qa = pipeline("question-answering")
```

```python
qa(context="My name is Naji Aboo and I work at Triassic Solutions", question="Where do Naji Aboo works?")
```
